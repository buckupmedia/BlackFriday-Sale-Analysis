## GENDER ANALYSIS

```python
# Group the DataFrame 'df' by the 'Gender' column
df_gender = df.groupby("Gender")
```

```python
# Create a figure and a set of subplots with 1 row and 2 columns
fig, axs = plt.subplots(1, 2, figsize=(10, 3))

# Plotting a pie chart to visualize the gender ratio
# 'explode' parameter is used to highlight the second slice of the pie
explode = [0, 0.08]
df_gender.size().plot(kind="pie", autopct="%.1f", ax=axs[0], explode=explode, title="GENDER RATIO")

# Plotting a bar graph to visualize the gender ratio
# The color parameter is set to differentiate between the two bars
df_gender.size().plot(kind="bar", ax=axs[1], title="GENDER RATIO", color=["pink", "skyblue"],   edgecolor  = "black", linewidth = .75)

# Adjusting the space between the subplots
plt.subplots_adjust(wspace=3)

# Display the plots
plt.show()
```

# *Lets find the purchases of each gender*

```python
# Group the DataFrame 'df' by the 'Gender' column
# For each group, calculate the sum of the 'Purchase' column
# Plot the resulting sums as a pie chart
# The 'autopct' parameter formats the percentage display to one decimal place
df.groupby("Gender")["Purchase"].sum().plot(kind = "pie", autopct = "%.2f")
```
